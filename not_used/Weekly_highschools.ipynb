```python
import pandas as pd
import os
from glob import glob
from natsort import natsorted
import geopandas as gpd
from tqdm import tqdm
```

```python
csv_dir = r'D:\Data\Advan\Weekly_Patterns\2022'
```

```python
school_key_fname = r"D:\OneDrive_Emory\OneDrive - Emory\Research_doc\hourly_population\data\highschool_school_placekey.csv"

school_key_df = pd.read_csv(school_key_fname)
school_placekeys = school_key_df['Placekey'].to_list()
print("Loaded school placekeys:", len(school_placekeys))
school_key_df
```

```python
save_dir = r'D:\Data\Advan\Weekly_Patterns\extracted_schools_2022'
```

```python
all_files = glob(os.path.join(csv_dir, '*.csv.gz'))
all_files = natsorted(all_files)
print(f"Found {len(all_files)} weekly pattern files.")
```

```python
df_list = []
for file in tqdm(all_files):
    week_df = pd.read_csv(file, dtype={'placekey': str})
    school_week_df = week_df[week_df['placekey'].isin(school_placekeys)]
    if not school_week_df.empty:
        week_fname = os.path.basename(file).replace('.csv.gz', '_schools.csv.gz')
        save_path = os.path.join(save_dir, week_fname)
        school_week_df.to_csv(save_path, index=False)
        df_list.append(school_week_df)

final_df = pd.concat(df_list, ignore_index=True)
final_df.to_csv(os.path.join(save_dir, 'all_weeks_highschools_2022.csv'), index=False)
```

```python
# save final_df to parquet
final_df.to_parquet(os.path.join(save_dir, 'all_weeks_highschools_2022.parquet'), index=False)
```

```python
import duckdb

# Connect to DuckDB (in-memory, or specify a file db)
con = duckdb.connect()
con.execute("PRAGMA threads=8;")   # use multiple cores
con.execute("PRAGMA memory_limit='18GB';")        # cap RAM usage at 18 GB
 
# Create a temp table of placekeys
con.execute("CREATE TABLE filter_keys(placekey VARCHAR);")
con.executemany("INSERT INTO filter_keys VALUES (?)", [(pk,) for pk in school_placekeys])

# Read all CSV.gz files, filter by placekeys, and write to Parquet
query = """
COPY (
    SELECT f.*
    FROM read_csv_auto('D:/Data/Advan/Weekly_Patterns/2022/*.csv.gz', header=TRUE) f
    INNER JOIN filter_keys k
    ON f.placekey = k.placekey
) TO 'D:/Data/Advan/Weekly_Patterns/filtered_patterns.parquet' (FORMAT PARQUET);
"""
con.execute(query)

print("✅ Filtering complete. Results saved to Parquet.")
```
