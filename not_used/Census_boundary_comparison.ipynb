```python
import geopandas as gpd
from tqdm import tqdm
import os
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
ACS2019_identical_to_2020_fname = r"D:\OneDrive_PSU\OneDrive - The Pennsylvania State University\Research_doc\Wild_fire\ACS_2019_CBG_identical_2020.csv"
ACS2020_identical_to_2019_fname = r"D:\OneDrive_PSU\OneDrive - The Pennsylvania State University\Research_doc\Wild_fire\ACS_2020_CBG_identical_2019.csv"

```

```python
ACS2019_identical_to_2020_df = pd.read_csv(ACS2019_identical_to_2020_fname).sort_values('GEOID')
ACS2019_identical_to_2020_df['CBG_2019'] = ACS2019_identical_to_2020_df['GEOID'].astype(str).str.zfill(12)
ACS2019_identical_to_2020_df
```

```python
ACS2020_identical_to_2019_df = pd.read_csv(ACS2020_identical_to_2019_fname).sort_values('GEOID')
ACS2020_identical_to_2019_df['CBG_2020'] = ACS2020_identical_to_2019_df['GEOID'].astype(str).str.zfill(12)
ACS2020_identical_to_2019_df
```

```python
merged_df = ACS2020_identical_to_2019_df.merge(ACS2019_identical_to_2020_df[['CBG_2019', 'ALAND', 'AWATER']], left_on='CBG_2020', right_on='CBG_2019', suffixes=('_2020', "_2019"))
merged_df
```

```python
merged_df['Land_dif'] = merged_df['ALAND_2020'] + merged_df['AWATER_2020'] - merged_df['ALAND_2019'] - merged_df['AWATER_2019']
merged_df['Land_dif'].sample(100).sum()
```
