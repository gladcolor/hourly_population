# Create database

```python
# get Safegarph patterns files

POI_files = glob.glob(r'J:\Safegraph\weekly_county_files\weekly_patterns_2018_2021\POI\*.csv')
POI_files = POI_files[::-1]
print(f"Found {len(POI_files)} files.")
POI_files[0]
```

```python
save_dir = r'J:\Safegraph\weekly_county_files\weekly_patterns_2018_2021\POI_sqlite'

for idx, file in enumerate(POI_files[:]):
    print(idx+1, file)
    basename = os.path.basename(file)
    df = pd.read_csv(file)
    new_name = os.path.join(save_dir, basename[:-3] + 'db')
    conn = sqlite3.connect(new_name)
    curs = conn.cursor()
    
    columns = ','.join(df.columns)
    columns.replace('placekey,parent_placekey', 'placekey PRIMARY KEY,parent_placekey')
    
    curs.execute('create table if not exists POI ' +
                f"({','.join(df.columns)})")
    df.to_sql('POI', conn, if_exists='replace', index=False)
    
    sql = f'CREATE INDEX placekey_idx ON POI(placekey);'    
    curs.execute(sql)
    
    conn.close
```
